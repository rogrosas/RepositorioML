**Carga de Datos**

```python
import pandas as pd
from pathlib import Path

ruta_base = Path(r"..\datos_examen")

application = pd.read_parquet(ruta_base / "application_.parquet")
bureau = pd.read_parquet(ruta_base / "bureau.parquet")
bureau_balance = pd.read_parquet(ruta_base / "bureau_balance.parquet")
previous = pd.read_parquet(ruta_base / "previous_application.parquet")
pos_cash = pd.read_parquet(ruta_base / "POS_CASH_balance.parquet")
installments = pd.read_parquet(ruta_base / "installments_payments.parquet")
credit_card = pd.read_parquet(ruta_base / "credit_card_balance.parquet")

# Validación de llaves
application['SK_ID_CURR'].is_unique
bureau['SK_ID_BUREAU'].is_unique
previous
```

```python
def resumen_nulos(df):
    return (df.isnull().mean().sort_values(ascending=False) * 100).head(20)

resumen_nulos(application)
resumen_nulos(bureau)
resumen_nulos(previous)
resumen_nulos(pos_cash)
resumen_nulos(installments)
resumen_nulos(credit_card)
```
